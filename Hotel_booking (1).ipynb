## EDA ---> Exploatory Data analysis

```python
# Checking percentage of canceled and not cancled
cancelled_perc = df['is_canceled'].value_counts(normalize= True)
cancelled_perc 
```

```python
# Barplot
plt.figure(figsize =(5,4))
plt.title("Reservation Status Count")
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts(),width = 0.7)
plt.show()
```

### 📊 **Insights from Reservation Status Count**

1. **Majority of Bookings Are Not Canceled**

   * The first bar (Not Canceled) is significantly taller.
   * This indicates that **most customers follow through with their reservations**.

2. **But Cancellation Rate Is Still Significant**

   * The second bar (Canceled) is not negligible — it's around **35-40% of total bookings** (visually estimated).
   * This means cancellations are a **major factor impacting revenue and occupancy**.



```python
# Countplot
ax1 = sns.countplot( x='hotel', hue = 'is_canceled', data = df)
legends_labels,_= ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor= (1,1))
plt.xlabel('Hotel')
plt.ylabel('number of reservation')
plt.legend(['Not Canceled','Canceled'])
plt.show()
```

### 📊 **Insights from the Hotel-Wise Reservation Status**

1. **City Hotels Have More Bookings Overall**

   * Both canceled and not-canceled bookings are higher for City Hotels.
   * This means City Hotels are generally more popular / have higher demand.

2. **Cancellation Rate Is Higher for City Hotels**

   * The orange bar (Canceled) is much closer to the blue bar for City Hotels than for Resort Hotels.
   * This suggests **a higher cancellation percentage in City Hotels**.

3. **Resort Hotels Have Lower Cancellations**

   * The gap between Not Canceled and Canceled bookings is larger for Resort Hotels.
   * This means Resort Hotel customers are more likely to complete their bookings.

```python
resort_hotel = df[df['hotel']== 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize= True)
```

```python
city_hotel = df[df['hotel']== 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize= True)
```

```python
# Grouping of resort_gotel and city_hotel in reservation status date
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# line Graph
plt.figure(figsize= (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize= 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label= 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label= 'City Hotel')
plt.legend(fontsize= 15)
plt.show()
```

### 📊 **Insights from ADR Trend Chart**

1. **Seasonality in Prices**

   * Both lines (Resort and City Hotels) show **cyclical patterns** — prices rise and fall at regular intervals.
   * Peaks are likely during **holiday seasons (summer & year-end)**, when demand is high.
   * Dips correspond to **off-season months** when demand is lower.

2. **Resort Hotel Prices Are More Volatile**

   * The blue line (Resort Hotels) fluctuates much more sharply compared to City Hotels.
   * This indicates that **Resort Hotels adjust prices dynamically** based on demand (holidays/weekends vs off-season).

3. **City Hotels Have a More Stable ADR**

   * The orange line is smoother and stays within a narrower range.
   * City Hotels likely cater to **business travelers** and have more consistent demand year-round.

4. **Overall Price Levels**

   * City Hotels (orange) generally maintain **higher average ADR** compared to Resort Hotels (blue), especially in the middle portion of the timeline.
   * But during peak seasons, Resort Hotels occasionally **surpass City Hotel prices**.



```python
# Countplot
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize= (16,8))
ax1 = sns.countplot(x= 'month', hue= 'is_canceled', data= df, palette= 'bright')
legends_labels,_= ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor= (1,1))
plt.title('Reservation Status per month', size= 20)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['Not Canceled','Canceled'])
plt.show()
```

### Insights
Not Canceled Reservations Dominate: In every month, the number of "Not Canceled" reservations (blue bars) is consistently higher than "Canceled" reservations (orange bars), indicating that most reservations are honored throughout the year.

Peak in Summer Months: The highest number of reservations occurs in July and August (months 7 and 8), with July having the single highest count. Notably, the gap between not canceled and canceled reservations is widest in these months, suggesting strong demand and better retention of bookings during peak vacation periods.

Cancellation Trend: While cancellations remain lower than completed bookings in all months, there is a noticeable reduction in cancellations from January to August, followed by a slight increase in October. The lowest cancellation rates occur during July, August, and September, coinciding with peak booking months.

Seasonal Patterns: Months 1 (January), 2, and 12 (December) show relatively higher cancellation rates compared to the mid-year period, possibly due to seasonal fluctuations, travel uncertainties, or changing consumer behavior in off-peak months.

Steady Mid-Year Bookings: From March to October, the number of reservations (both canceled and not canceled) remains relatively steady, with a noticeable dip in cancellations over the summer (months 6 to 9).

```python
# Barplot
plt.figure(figsize= (15,8))
plt.title('Avg Daily Rate(ADR) per month', fontsize= 30)
sns.barplot(x='month', y= 'adr', data= df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

### Insights
Highest ADR in January: January (month 1) has the highest ADR, exceeding 550,000, indicating that rooms are priced at a premium at the start of the year, potentially due to holiday or peak demand periods.

Steady Decline Until August: ADR generally decreases from January through August, with a small uptick in July (month 7), suggesting mid-year discounts or a response to increased competition and changing demand dynamics.

Second Peak in July: July shows a notable spike in ADR, which may be attributed to high demand during summer vacation, allowing hotels to charge more per room.

Lowest in September and November: September (month 9) and November (month 11) have the lowest ADR values, possibly reflecting lower tourism activity or strategic price reductions to attract more guests during these months.

End-of-Year Recovery: December (month 12) experiences a slight increase, which could be driven by holiday travel and events, but not reaching the heights observed in January or July.

```python
# Pie Chart
canceled_data = df[df['is_canceled']==1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize= (8,8))
plt.title('Top 10 Countries with Reservation canceled')
plt.pie(top_10_country, autopct= '%.2f', labels= top_10_country.index)
plt.show()
```

### Insights
Portugal (PRT) Dominates Cancellations: The vast majority of canceled reservations come from Portugal, representing 70.13% of the total. This is a striking concentration, suggesting unique factors—such as local guest behavior, promotional policies, or market expectations—heavily influence cancellations within Portugal.

Significant Contribution from UK and Spain: The United Kingdom (GBR) and Spain (ESP) stand out as the next most frequent sources, at 6.24% and 5.53% respectively. France (FRA) and Italy (ITA) also contribute notable shares, 4.92% and 3.39%.

Long Tail of Other Countries: The remaining countries (Germany, Ireland, Brazil, USA, and Belgium) each contribute between about 1% and 3% of cancellations, indicating diverse but much smaller cancellation activity outside the leading countries.
